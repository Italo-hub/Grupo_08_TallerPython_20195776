```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom
```

```python
import math
import statistics
from scipy import stats
from scipy.stats import norm
```

### Poblema 1

```python
moneda = binom.rvs(1, 0.5, size=10)
moneda
```

```python
probabilidad_7caras = binom.pmf(7, 10, 0.5)

print(f"La probabilidad de obtener 7 caras en 10 lanzamientos es de {probabilidad_7caras:.4f}")
```

### Poblema 2

```python
prob_acumulada_moneda = binom.cdf(7, 10, 0.5)

print(f"La probabilidad de obtener 7 o menos caras en 10 lanzamientos es de {prob_acumulada_moneda:.4f}")
```

### Poblema 3

```python
# Para obtener los valores de 0 a 10
valores = np.arange (0, 11)

# Probabilidades para cada valor
probabilidades = binom.pmf( valores, 10, 0.5)

plt.bar(valores, probabilidades, color='skyblue')
plt.xlabel('Número de Caras')
plt.ylabel('Probabilidad')
plt.title('Distribución Binomial: 10 lanzamientos de una moneda sin trucos')

plt.show()
```

### Poblema 5

```python
from scipy.stats import norm

media = 0
desviacion_estandar = 1
datos_1 = norm.rvs(loc=media, scale= desviacion_estandar, size= 1000)

```

### Poblema 6

```python
media_muestral =np.mean(datos_1)
desviacion_muestral = np.std(datos_1)
mediana_muestral = np.median(datos_1)
```

```python
print(f"la media muestral es: {media_muestral}")
print(f"la desviación muestral es: {desviacion_muestral}")
print(f"la mediana muestral es: {mediana_muestral}")
```

### Poblema 7

```python
plt.hist(datos_1, bins=30, density=True, alpha=0.7, color= 'blue', edgecolor= 'purple' )
x = np.linspace(-3, 3, 1000)
pdf_teorica = norm.pdf(x, loc = media, scale=desviacion_estandar)
plt.plot(x, pdf_teorica, 'k', lw=2, label='Distribución Normal Teórica' )
plt.xlabel('valor')
plt.ylabel('Densidad de probabilidad')
plt.title('Histograma de datos generados y curva de distribución normal')
plt.legend()
plt.show()
```

### Poblema 8

```python
probabilidad_ocurrencia = norm.cdf(1, loc= media, scale=desviacion_estandar) - norm.cdf(-1,loc= media, scale= desviacion_estandar)
```

```python
print(f"la probabilidad de ocurrencia de que el valor caiga entre el -1 y 1, es de: {probabilidad_ocurrencia}")
```
